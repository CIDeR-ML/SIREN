```python
import numpy as np
import torch
import h5py

from iotools import *
from iotools.sampler import *
import iotools.Dataset as dt
import iotools.loader_factory as lf

filename = '/home/pdeperio/wcte/ryom/fiTQun_tuning/TuningFiles/scattable/fiTQun_scattablesF_nuPRISMBeamTest_16cShort_mPMT.h5'
region = 'top'
num_samples = 100
start_frac = 0.1
dset = dt.WCTEScatTable(h5_file = filename, region = region, num_samples_per_load = num_samples)

rnd_loader = lf.loader_factory('WCTEScatTable', batch_size=1, shuffle=True, 
                               sampler = rndSampleWCTE(dset),
                               num_workers=0, pin_memory = True, h5_file = filename, region = region, num_samples_per_load = num_samples)
seq_loader = lf.loader_factory('WCTEScatTable', batch_size=1, shuffle=True, 
                               sampler = seqSampleWCTE(dset, start_frac),
                               num_workers=0, pin_memory = True, h5_file = filename, region = region, num_samples_per_load = num_samples)
```

```python
bins_rnd, data_rnd = next(iter(rnd_loader))
print(f"Randomly sampled bin:{bins_rnd}, data:{data_rnd}")
bins_seq, data_seq = next(iter(seq_loader))
print(f"Sequntially sampled from {start_frac*100}% of the full dataset - bin:{bins_seq}, data:{data_seq}.")
```
